```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# Assigning numbers to the words
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
# splitting the data on the basis of next line to form sentences
# and converting in a supervised learning form so that they can be written in the form of input and output
input_sequence = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequence.append(tokenized_sentence[:i+1])
```

```python
input_sequence
```

```python
# now applying zero padding so that the length of all the input_sequence becomes same
max([len(x) for x in input_sequence])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequence = pad_sequences(input_sequence,maxlen = 261,padding = 'pre')
```

```python
X = padded_input_sequence[:,:-1]
```

```python
y = padded_input_sequence[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
# this should be considered as classification problem as in regression it will give output as 2.1 that doesn't represent any word
# so we will use multiclass classification and onehot encode the output of the length of total unique words
from tensorflow.keras.utils import to_categorical
vocab_size = len(tokenizer.word_index) + 1
y = to_categorical(y,num_classes = vocab_size)
```

```python
y.shape
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,LSTM
```

```python
model = Sequential()
model.add(Embedding(vocab_size,100, input_shape=(X.shape[1],)))
model.add(LSTM(150))
model.add(Dense(vocab_size,activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs = 30)
```

```python
import numpy as np
# to predict
text = "I can confess that my"
#tokenize
token_text = tokenizer.texts_to_sequences([text])[0]
# padding
padded_token_text = pad_sequences([token_text],maxlen =57,padding='pre')
# predict
model.predict(padded_token_text)
pos = np.argmax(model.predict(padded_token_text))

for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
```
